```python
import pandas as pd

df_aud = pd.read_csv('Данные об аудитории.csv', encoding='utf-8')

mau = df_aud['user_id'].nunique()
print(f"Ответ на вопрос №1 (MAU): {mau}")
```

```python
df_aud['date'] = pd.to_datetime(df_aud['date'])

daily_unique = df_aud.groupby('date')['user_id'].nunique()

mean_dau = daily_unique.mean()

print(f"Средний DAU: {mean_dau:.0f}")
```

```python
users_1nov = set(df_aud[df_aud['date'] == '2023-11-01']['user_id'])

users_2nov = set(df_aud[df_aud['date'] == '2023-11-02']['user_id'])

returned = users_1nov.intersection(users_2nov)

retention_d1 = (len(returned) / len(users_1nov)) * 100

print(f"Кол-во пользователей 1.11: {len(users_1nov)}")
print(f"Вернулись 2.11: {len(returned)}")
print(f"Retention D1: {retention_d1:.2f}%")
```

```python
users_with_views = df_aud[df_aud['view_adverts'] > 0]['user_id'].nunique()

conversion = (users_with_views / mau) * 100

print(f"Пользователей с просмотрами: {users_with_views}")
print(f"Общее кол-во пользователей (MAU): {mau}")
print(f"Конверсия в просмотр: {conversion:.1f}%")
```

```python
import pandas as pd
from scipy import stats

try:
    df_ab = pd.read_csv('Данные АБ тестов.csv', sep=None, engine='python', encoding='utf-8-sig')
except:
    df_ab = pd.read_csv('Данные АБ тестов.csv', sep=None, engine='python', encoding='cp1251')

df_ab.columns = df_ab.columns.str.replace('﻿', '').str.strip()

def analyze_test(num):
    control = df_ab[(df_ab['experiment_num'] == num) & (df_ab['experiment_group'] == 'control')]['revenue']
    test = df_ab[(df_ab['experiment_num'] == num) & (df_ab['experiment_group'] == 'test')]['revenue']
    
    t_stat, p_val = stats.ttest_ind(control, test)
    
    print(f"--- Тест №{num} ---")
    print(f"ARPU Контроль: {control.mean():.2f}")
    print(f"ARPU Тест: {test.mean():.2f}")
    print(f"p-value: {p_val:.4f}")
    
    if p_val < 0.05:
        print("Результат: Различия СТАТИСТИЧЕСКИ ЗНАЧИМЫ")
    else:
        print("Результат: Различия НЕЗНАЧИМЫ")
    print("-" * 20)

for i in [1, 2, 3]:
    analyze_test(i)
```

```python
import pandas as pd

try:
    df_listers = pd.read_csv('Листеры.csv', sep=None, engine='python', encoding='utf-8-sig')
except:
    df_listers = pd.read_csv('Листеры.csv', sep=None, engine='python', encoding='cp1251')

total_revenue_listers = df_listers['revenue'].sum()
unique_listers = df_listers['user_id'].nunique()

arpu_listers = total_revenue_listers / unique_listers

print(f"Общая выручка: {total_revenue_listers}")
print(f"Количество уникальных листеров: {unique_listers}")
print(f"Средний доход на листера: {arpu_listers:.1f}")
```
